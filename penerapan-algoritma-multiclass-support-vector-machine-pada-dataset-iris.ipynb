```python
# baca data
import pandas as pd
url=("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
headers=["sepal_length", "sepal_width", "petal_length", "petal_width", "class_label"]
datairis = pd.read_csv(url, names=headers)
print("===============================================================================")
print("                                  Data head")
print("===============================================================================")
print(datairis.head())

#pisah fitur dengan kelas
import numpy as np
fitur=np.array(datairis[["sepal_length", "sepal_width", "petal_length", "petal_width"]])
kelas=np.array(datairis["class_label"])

# import scv dan training model
from sklearn.svm import SVC
modeliris = SVC(decision_function_shape='ovo')
modeliris.fit(fitur, kelas)

# test model
print("\n\n==================================================")
print("                    Test Model")
print("==================================================")
print("Kelas untuk data [5,3,1.6,0.3] :", modeliris.predict([[5,3,1.6,0.3]]))

# melihat nilai mana saja yang menjadi support vector
print("\n\n=====================================================")
print(" Melihat model mana saja yang menjadi support vector")
print("=====================================================")
print(modeliris.support_vectors_)

# data index baris ke berapa saja yang menjadi support vector
print("\n\n=====================================================================")
print(" Melihat data index baris ke berapa saja yang menjadi support vector")
print("=====================================================================")
print(modeliris.support_)

# melihat banyaknya data yg menjadi support vector dalam setiap kelas
print("\n\n=====================================================================")
print(" Melihat banyaknya data yg menjadi support vector dalam setiap kelas")
print("=====================================================================")
print(modeliris.n_support_)
```
